```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
T = np.linspace(-20,20,90)   # here values change like 90, or 10 or 80 to increase curve resolution and 
Y = sigmoid(T)
plt.plot(T, Y) # T is axis and Y on y axis
```
