```python
from keras.utils import plot_model
from keras.models import Model
from keras.layers import Input
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers.convolutional import Conv2D
from keras.layers.pooling import MaxPooling2D
from keras.layers.merge import concatenate
from IPython.display import SVG
from keras.utils.vis_utils import model_to_dot
from IPython.display import SVG
from keras.utils.vis_utils import model_to_dot
```

```python
# input layer
visible = Input(shape=(64,64,1))
# first feature extractor
conv1 = Conv2D(32, kernel_size=4, activation='relu')(visible)
pool1 = MaxPooling2D(pool_size=(2, 2))(conv1)
flat1 = Flatten()(pool1)
#second feature extractor
conv2 = Conv2D(16, kernel_size=8, activation='relu')(visible)
pool2 = MaxPooling2D(pool_size=(2, 2))(conv2)
flat2 = Flatten()(pool2)

#THIRD FEATURE EXTRACTOR
conv3 = Conv2D(8, kernel_size=16, activation='relu')(visible)
pool3 = MaxPooling2D(pool_size=(2, 2))(conv3)
flat3 = Flatten()(pool3)

#FOURTH FEATURE EXTRACTOR
conv4 = Conv2D(8, kernel_size=16, activation='relu')(visible)
pool4 = MaxPooling2D(pool_size=(2, 2))(conv4)
flat4 = Flatten()(pool4)

#FIFTH FEATURE EXTRACTOR
conv5 = Conv2D(8, kernel_size=16, activation='relu')(visible)
pool5 = MaxPooling2D(pool_size=(2, 2))(conv5)
flat5 = Flatten()(pool5)

# merge feature extractors
merge = concatenate([flat1, flat2,flat3,flat4, flat5])
# interpretation layer
hidden1 = Dense(10, activation='relu')(merge)
# prediction output
output = Dense(1, activation='sigmoid')(hidden1)
model = Model(inputs=visible, outputs=output)
```

```python
# plot graph
plot_model(model, to_file='shared_input_layer.png')
```

```python
SVG(model_to_dot(model, show_shapes=True).create(prog='dot', format='svg'))
```
