```python
inputs = Input(shape=(32,32,3),name='X')
```

```python
y = Conv2D(3,(3,3), padding='same',name='WEIGHT_LAYER_1')(inputs)
y1=Activation('relu',name='RELU1')(y)
```

```python
y2 = Conv2D(3,(3,3), padding='same',name='WEIGHT_LAYER_2')(y1)
```

```python
z = keras.layers.add([inputs,y2],name='INDENTITY_X')
```

```python
output=Activation('relu',name='RELU2')(z)
```

```python
model=Model(inputs,output)
```

```python
model.summary()
```

```python
SVG(model_to_dot(model, show_shapes=True).create(prog='dot', format='svg'))
```
