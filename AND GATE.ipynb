```python
#TRAIN AND TEST MODEL ON AND GATE
import keras
import numpy as np
```

```python
input = np.array([[0,0],
                  [0,1],
                  [1,0],
                  [1,1]])
print(input)
```

```python
output = np.array([[0],
                     [0],
                     [0],
                     [1]])
print(output)
```

```python
model = keras.models.Sequential(layers=[ 
    keras.layers.Dense(input_dim =2, units = 1),
    keras.layers.Activation(keras.activations.sigmoid)

])


#model.compile(optimizer=keras.optimizers.SGD(lr=.5),loss='mse')
```

```python
model.summary()
```

```python
model.compile(optimizer=keras.optimizers.SGD(lr=.5),loss='mse')
```

```python
model.weights # shape (2,1) = 2 input neurons 1 output neurons
```

```python
keras.backend.eval(model.weights[1])
```

```python
model.fit(input, output, epochs = 1000) # if you use , verbose = False than iteration can not show
```

```python
model.predict(np.array([[0,1]]))
```

```python
model.predict(np.array([[1,1]]))
```
