```python
import numpy as np #numpy is used for array operation
import matplotlib.pyplot as plt
import os
import cv2
```

```python
DATADIR = "/home/wazir/kagglecatsanddogs_3367a/PetImages"
CATEGORIES = ["Dog","Cat"]
for category in CATEGORIES:
    path = os.path.join(DATADIR,category)
    for img in os.listdir(path):
        img_array=cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE) #HERE SHOW RGB
        #img_array=cv2.imread(os.path.join(path,img)) #IF WE CAN NOT USE cv2.IMREAD_GRAYSCALE OR ABOVENOW SHOW AS BGR
        plt.imshow(img_array,cmap="gray")
        plt.show()
        break
    break
```

```python
#SOME IMAGE IN DIFFERENT SHAPE PHOTOS, SOME In LANDSCAPE, SQUARE SIZE , SOME OTHER STYLE SO WE MAKE 
#AS NORMALIZE , SO WE MAKE AS SAME SHAPE
#SO WE DECIDE EVERY IMAGE HAS SIZE 50x50
IMG_SIZE =50
#WAY IS TO CREATE NEW ARRAY
new_array = cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
plt.imshow(new_array,cmap='gray')
plt.show()
```

```python
print(new_array.shape)
```

```python
training_data = [] #Create training_dat = to empty list
def create_training_data():
    for category in CATEGORIES:
        path = os.path.join(DATADIR,category)
        class_num = CATEGORIES.index(category) #HERE WE DEFINE 0 is Dog or 1 is Cat
        for img in os.listdir(path):
            try:
                img_array=cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE) #HERE SHOW RGB
                new_array = cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
                training_data.append([new_array,class_num])
            except Exception as e:
                pass
create_training_data()
```

```python
print(len(training_data))
```

```python
#NOW WE SHUFFLE TRAINING DATA
import random
random.shuffle(training_data[:10]) #SHUFFLING DATA
```

```python
x=[]
y=[] #CREATE TWO TARINING AND VALIDATAION EMPTY LIST
```

```python
for features, label in training_data:
    x.append(features)
    y.append(label)
x = np.array(x).reshape(-1,IMG_SIZE,IMG_SIZE,1) #1 used for Gray Scale
```

```python
import pickle
pickle_out=open("x.pickle","wb")
pickle.dump(x,pickle_out)
pickle_out.close()

pickle_out=open("y.pickle","wb")
pickle.dump(x,pickle_out)
pickle_out.close()
```
