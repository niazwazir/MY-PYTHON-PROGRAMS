# METHOD-1

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = Image.open("./zebra.bmp")
```

```python
fig = plt.figure(figsize=(18, 16), dpi= 100)
#fig = plt.figure(figsize=(18, 16), dpi= 255)
ax = plt.subplot("131")
ax.imshow(img)
plt.axis('off')
```
