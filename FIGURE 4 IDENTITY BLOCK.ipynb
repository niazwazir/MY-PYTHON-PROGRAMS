```python
inputs = Input(shape=(32,32,1),name='Input') #A THIS PART OF CHANNEL 1 ISAME AS 
c = Conv2D(3,(3,3), padding='same',name='Convolution1')(inputs)
b = BatchNormalization(axis = 3, name = 'Batch_Norm_1')(c)
r = Activation('relu',name='ReLU1')(b)

c1 = Conv2D(3,(3,3), padding='same',name='Convolution2')(r)
b1 = BatchNormalization(axis = 3, name = 'Batch_Norm_2')(c1)
r1 = Activation('relu',name='ReLU2')(b1)

c2 = Conv2D(3,(3,3), padding='same',name='Convolution3')(r1)
b2 = BatchNormalization(axis = 3, name = 'Batch_Norm_3')(c2)

z = keras.layers.add([inputs,b2],name='Addition')

output=Activation('relu',name='ReLU3')(z)


model=Model(inputs,output)
```
