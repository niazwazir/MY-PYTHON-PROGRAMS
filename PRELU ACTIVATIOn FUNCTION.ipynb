```python
import keras
import cv2
from keras.models import Sequential
from keras.layers import Dense, Conv2D, Activation, Convolution2D
import numpy as np
from keras.layers.advanced_activations import PReLU
import matplotlib.pyplot as plt
```

```python
baby= cv2.imread('Set5/baby_GT.bmp')
```

```python
baby.shape
```

```python
def nice_baby_printer(model,baby):
    baby_batch = np.expand_dims(baby,axis=0)
    conv_baby2 = model.predict(baby_batch)
    
    conv_baby2 = np.squeeze(conv_baby2, axis=0)
    print (conv_baby2.shape)
    conv_baby2 = conv_baby2.reshape(conv_baby2.shape[:2])
    
    print (conv_baby2.shape)
    plt.imshow(conv_baby2)
```

```python
model = Sequential()
act = keras.layers.advanced_activations.PReLU(init='zero', weights=None)
model.add(Convolution2D(1,3,3,input_shape=baby.shape))
model.add(act)
nice_baby_printer(model,baby)
```
