```python
%matplotlib inline
import numpy as np
import cv2
import matplotlib.pyplot as plt
from PIL import Image
import time, math
%matplotlib inline

S = 8 #UPSCALING FACTOR
```

```python
# READ THE GROUND TRUTH REFERENCE IMAGE
ref = cv2.imread("/home/wazir/SRCNN MODEL/SRCNN/Test/Set5/baby_GT.bmp",0)
ref = cv2.normalize(ref.astype('float'), None, 0.0, 1.0, cv2.NORM_MINMAX)


#CREATE LOW RESOLUTION IMAGE FROM REFERENCE IMAGE
LRImage = cv2.resize(ref, (int(ref.shape[1]/S), int(ref.shape[0]/S)))


#CREATE HIGH RESOLUTION IMAGE USING BICUBIC INTERPOLATION
LRImage1 = cv2.resize(LRImage,(ref.shape[1],ref.shape[0]),cv2.INTER_CUBIC)

fig = plt.figure(figsize = (18,16), dpi = 80)
ax = plt.subplot("131")
ax.imshow(ref, cmap='gray')
ax.set_title("GT")


ax = plt.subplot("132")
ax.imshow(LRImage, cmap='gray')
ax.set_title("LOW RESO IMAGE")



ax = plt.subplot("133")
ax.imshow(LRImage1, cmap = 'gray')
ax.set_title("HR BICUBIC")
```

```python
#CREATE HIGH RESOLUTION IMAGE USING INTERPOLATION TECHNIQUE
"""
interpolation 

interpolation method:

    INTER_NEAREST - a nearest-neighbor interpolation
    INTER_LINEAR - a bilinear interpolation (used by default)
    INTER_AREA - resampling using pixel area relation. It may be a preferred method for image decimation, 
    as it gives moire-free results. But when the image is zoomed, it is similar to the INTER_NEAREST method.
    INTER_CUBIC - a bicubic interpolation over 4x4 pixel neighborhood
    INTER_LANCZOS4 - a Lanczos interpolation over 8x8 pixel neighborhood
"""
#LRImage = cv2.resize(LRImage,(ref.shape[1],ref.shape[0]),cv2.INTER_LINEAR) 
#LRImage = cv2.resize(LRImage,(ref.shape[1],ref.shape[0]),cv2.INTER_NEAREST)
LRImage = cv2.resize(LRImage,(ref.shape[1],ref.shape[0]),cv2.INTER_CUBIC)
#LRImage = cv2.resize(LRImage,(ref.shape[1],ref.shape[0]),cv2.INTER_AREA)
#LRImage = cv2.resize(LRImage,(ref.shape[1],ref.shape[0]),cv2.INTER_LANCZOS4)
cv2.imshow('HIGH RESOLUTION IMAGE USING BICUBIC INTERPOLATION',LRImage)
cv2.waitKey(0)
cv2.destroyAllWindows()
plt.imshow(LRImage)
plt.show()
```
