```python
#COMMAND PLT.IMSHOW THIS SHOW IMAGE AND CMAP = GRAY
#YOU CAN SHOW COLOR BECAUSE OPENCV USED BGR AND MATPLOT LIB
#USED RGB, AND USED INTERPOLATION AS BICUBIC
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#img = cv2.imread('zebra.bmp')
img = cv2.imread('zebra.bmp', cv2.IMREAD_GRAYSCALE)
```

```python
#plt.imshow(img) # BECAUSE plt show RGB
#plt.imshow(img, cmap='gray', interpolation='bicubic')
plt.imshow(img, cmap='gray', interpolation='nearest')
plt.plot([50,100],[80,100],'r',linewidth=5)
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
import numpy as np
```

```python
im = np.array(Image.open('zebra.bmp'), dtype=np.uint8)
# Create figure and axes
fig,ax = plt.subplots(1)
# Display the image
ax.imshow(im)
# Create a Rectangle patch
rect = patches.Rectangle((400, 100),80,70,linewidth=5,edgecolor='r',facecolor='none')
# Add the patch to the Axes
ax.add_patch(rect)
plt.show()
```
