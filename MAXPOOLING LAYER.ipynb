```python
from keras.models import Sequential
from keras.layers import Dense, Conv2D, Activation
import keras.backend as K
from keras import optimizers
import numpy as np
import h5py
from keras.layers.advanced_activations import PReLU
from keras.layers import Convolution2D
import cv2
import matplotlib as plt
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from keras.layers import Conv2D, MaxPooling2D, Input, Merge, ZeroPadding2D, merge
```

```python
baby= cv2.imread('Set5/baby_GT.bmp')
```

```python
baby.shape
```

```python
def nice_baby_printer(model,baby):
    baby_batch = np.expand_dims(baby,axis=0)
    conv_baby2 = model.predict(baby_batch)
    
    conv_baby2 = np.squeeze(conv_baby2, axis=0)
    print (conv_baby2.shape)
    conv_baby2 = conv_baby2.reshape(conv_baby2.shape[:2])
    
    print (conv_baby2.shape)
    plt.imshow(conv_baby2)
```

```python
model = Sequential()
model.add(Convolution2D(1,3,3,input_shape=baby.shape))
nice_baby_printer(model,baby)
```

```python
model = Sequential()
model.add(Convolution2D(1,3,3,input_shape=baby.shape))
model.add(Activation('relu'))
nice_baby_printer(model,baby)
```

```python
model = Sequential()
model.add(Convolution2D(1,3,3,input_shape=baby.shape))
model.add(MaxPooling2D(pool_size=(2,2)))
nice_baby_printer(model,baby)
```

```python
model = Sequential()
model.add(Convolution2D(1,3,3,input_shape=baby.shape))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
nice_baby_printer(model,baby)
```

```python
model = Sequential()
model.add(Convolution2D(1,3,3,input_shape=baby.shape))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(5,5)))
nice_baby_printer(model,baby)
```

```python
model = Sequential()
model.add(Convolution2D(1,3,3,input_shape=baby.shape))
model.add(Activation('relu'))
nice_baby_printer(model,baby)
```

```python
import keras
model = Sequential()
act = keras.layers.advanced_activations.PReLU(init='zero', weights=None)
model.add(Convolution2D(1,3,3,input_shape=baby.shape))
model.add(act)
nice_baby_printer(model,baby)
```
