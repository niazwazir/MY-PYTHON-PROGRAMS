```python
from skimage.measure import structural_similarity as ssim
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
def mse(imageA, imageC):
    err = np.sum((imageA.astype("float") - imageC.astype("float")) ** 2)
    err /= float(imageA.shape[0] * imageA.shape[1])
    return err
```

```python
def compare_images(imageA, imageC, title):
    m = mse(imageA, imageC)
    s = ssim(imageA, imageC)
    
    # setup the figure
    fig = plt.figure(title)
    #plt.suptitle("MSE: %.2f, SSIM: %.2f" % (m, s))
    plt.suptitle("SSIM: %.4f" % (s))
    
    # show first image
    ax = fig.add_subplot(1, 2, 1)
    plt.imshow(imageA, cmap = plt.cm.gray)
    plt.axis("off")
    
    # show the second image
    ax = fig.add_subplot(1, 2, 2)
    plt.imshow(imageC, cmap = plt.cm.gray)
    plt.axis("off")
    
    # show the images
    plt.show()
```

```python
original = cv2.imread("coastguard.png") ###############################################################
#contrast = cv2.imread("imageB.jpg")
shopped = cv2.imread("MODEL_OUTPUT_IMAGE.bmp")

# convert the images to grayscale
original = cv2.cvtColor(original, cv2.COLOR_BGR2GRAY)
#contrast = cv2.cvtColor(contrast, cv2.COLOR_BGR2GRAY)
shopped = cv2.cvtColor(shopped, cv2.COLOR_BGR2GRAY)
```

```python
fig = plt.figure("Images")
images = ("Original", original), ("Photoshopped", shopped)
# loop over the images
for (i, (name, image)) in enumerate(images):
    ax = fig.add_subplot(1, 2, i + 1)
    ax.set_title(name)
    plt.imshow(image, cmap = plt.cm.gray)
    plt.axis("off")
```

```python
compare_images(original, original, "Original vs. Original")
#compare_images(original, contrast, "Original vs. Contrast")
compare_images(original, shopped, "Original vs. Photoshopped")
```
